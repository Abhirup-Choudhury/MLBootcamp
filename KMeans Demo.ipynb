```python
import numpy as np
import matplotlib.pyplot as plt
from MLLibrary import StandardizationScale,NormalizationScale,Utilities,KMeansClustering
```

```python
x_data = Utilities.load_data("data\Kmean_dataset.csv")
x_data.shape
```

```python
x_train,x_cv = Utilities.split_data(70,x_data)
scaler = StandardizationScale()
x_train_scaled = scaler.fit_transform(x_train)
x_cv_scaled = scaler.transform(x_cv)
scaler2 = NormalizationScale()
x_train_scaled = scaler2.fit_transform(x_train_scaled)
x_cv_scaled = scaler2.transform(x_cv_scaled)

print(x_train.shape)
print(x_cv.shape)
```

```python
costs = []
models = []
epochs = 20
K = np.arange(2,30)
for i,k in enumerate(K):
    models.append(KMeansClustering(x_train_scaled,epochs,k))
    costs.append(models[i].cost_history[-1])

plt.plot(np.arange(len(K)), costs)
plt.xticks(np.arange(2,30,2))
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Cluster Numbers vs Cost')
plt.show()
```
